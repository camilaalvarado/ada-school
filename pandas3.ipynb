```python
import pandas as pd
from datasets import load_dataset

# Cargar el dataset
dataset = load_dataset("mstz/heart_failure")
data = dataset["train"]

# Crear un DataFrame de Pandas
df = pd.DataFrame(data)

# Filtrar las filas donde las personas son fumadoras
df_fumadores = df.loc[df['is_smoker'] == 1]

# Agrupar por género y calcular la cantidad de fumadores en cada grupo
resultados = df_fumadores.groupby('is_male').size()

# Imprimir los resultados
print(f"hombres fumadores: {resultados[1]}")
print(f"mujeres fumadoras: {resultados[0]}")
```
