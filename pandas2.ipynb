```python
import pandas as pd
from datasets import load_dataset


dataset = load_dataset("mstz/heart_failure")
data = dataset["train"]


df = pd.DataFrame(data)

# Filtrar las filas donde las personas fallecieron
df_fallecidos = df.loc[df['is_dead'] == 1]

# Filtrar las filas donde las personas no fallecieron
df_no_fallecidos = df.loc[df['is_dead'] == 0]

# Mostrar las primeras filas de cada DataFrame
print("DataFrame de personas fallecidas:")
print(df_fallecidos.head())

print("\nDataFrame de personas no fallecidas:")
print(df_no_fallecidos.head())
```

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("mstz/heart_failure")
data = dataset["train"]

df = pd.DataFrame(data)

# Verificar tipos de datos en cada columna
tipos_de_datos = df.dtypes
print("Tipos de datos en cada columna:")
print(tipos_de_datos)

# Filtrar columnas numéricas en formato de cadena
columnas_numericas_en_cadena = df.select_dtypes(include=['object']).columns
print("Columnas numéricas en formato de cadena:")
print(columnas_numericas_en_cadena)

# Filtrar columnas no numéricas en formato de cadena
columnas_no_numericas_en_cadena = df.select_dtypes(exclude=['object']).columns
print("Columnas no numéricas en formato de cadena:")
print(columnas_no_numericas_en_cadena)

# Filtrar columnas con tipos de datos específicos
columnas_con_tipos_especificos = df.select_dtypes(include=['int64', 'float64']).columns
print("Columnas con tipos de datos específicos (enteros o flotantes):")
print(columnas_con_tipos_especificos)

# Continuar con el resto de tu código...
df_fallecidos = df.loc[df['is_dead'] == 1]
df_no_fallecidos = df.loc[df['is_dead'] == 0]
promedio_edad_fallecidos = df_fallecidos['age'].mean()
promedio_edad_no_fallecidos = df_no_fallecidos['age'].mean()

print(f"Promedio de edades de personas fallecidas: {promedio_edad_fallecidos}")
print(f"Promedio de edades de personas no fallecidas: {promedio_edad_no_fallecidos}")
```

```python
import pandas as pd
from datasets import load_dataset

# Cargar el dataset
dataset = load_dataset("mstz/heart_failure")
data = dataset["train"]

# Crear un DataFrame de Pandas
df = pd.DataFrame(data)

# Filtrar las filas donde las personas son fumadoras
df_fumadores = df.loc[df['is_smoker'] == 1]

# Agrupar por género y calcular la cantidad de fumadores en cada grupo
resultados = df_fumadores.groupby('is_male').size()

# Imprimir los resultados
print(f"hombres fumadores: {resultados[1]}")
print(f"mujeres fumadoras: {resultados[0]}")
```
